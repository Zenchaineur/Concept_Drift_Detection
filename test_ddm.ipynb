```python
from DDM import *
from dataset import *
import matplotlib.pyplot as plt
```

```python
X, y, targets = dataset('datasets/ELEC.csv')
pm = PredictionManager(X, y, targets)
error_rate_list, nb_errors = pm.run_all_steps()
print("Error rate :", nb_errors/len(X))
print("Accuracy :", 100*(1-nb_errors/len(X)))
plt.plot(error_rate_list, color = 'blue')
plt.show()
```
